```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Categorical Data Plots

Now let's discuss using seaborn to plot categorical data .there are few main plot types of this

1.boxplot
2.violinplot
3.stripplot
4.swarmplot
5.barplot
6.countplot

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(x = df['sex'],hue = df['smoker'])
```

```python
import numpy as np
sns.barplot(x = df['sex'],y = df['total_bill'],estimator=np.sum)
```

```python
sns.barplot(x = df['sex'],y = df['tip'])
```

```python
sns.boxplot(x = 'tip',y = 'day',data = df,palette = "rainbow")
```

```python
sns.violinplot(x = df['day'],y = df['total_bill'])
```

```python
sns.stripplot(x = df['day'],y = df['total_bill'])
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.violinplot(x = df['tip'],y = df['day'])
sns.swarmplot(x = 'tip',y = 'day',data = df)
```
