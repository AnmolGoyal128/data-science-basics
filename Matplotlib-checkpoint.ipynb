### Object Oriented way

Matplotlib allows the aspect ratio,DPI,and figur size to be specified when the Figue object is created.You can use the figsize and dpi keyword argument.

figsize is a tuple of a width and height of the figure in inches DPI is the dots-per-inch(pixel per inch) 

```python
fig = plt.figure()
```

```python
fig
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.1])
axis1.plot(X,Y)
axis2 = fig.add_axes([0.5,0.3,0.2,0.8])
axis2.plot(X,Y)
```

```python
fig = plt.figure(figsize=(12,2),dpi = 300)
axis1 = fig.add_axes([0.1,0.1,0.8,0.1])
axis1.plot(X,Y)
axis2 = fig.add_axes([0.5,0.3,0.2,0.8])
axis2.plot(X,Y)
```
